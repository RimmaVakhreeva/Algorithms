```python
# 1
length = 25
square = [[0]*length for i in range(length)]
for x in range(length):
    for y in range(length):
        if x < y:
            square[x][y] = '#'
        else:
            square[x][y] = '.'
        print(square[x][y], end=' ')
    print()
```
